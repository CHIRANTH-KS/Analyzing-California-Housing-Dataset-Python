```python
# 1. What is the average median income of the data set and check the distribution of data using appropriate plots. 
#Please explain the distribution of the plot.

import pandas as pd

# Load the dataset
df= pd.read_excel("housing.xlsx")
df.sample(12)
```

```python
# Calculate the average median income
Avg_Median_income=df['median_income'].mean()

# Print the average median income
print("The average median income of the data set is : ",Avg_Median_income)
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot a histogram of median income with 40 bins
plt.hist(df['median_income'], bins=40,edgecolor='black')

# add labels for the x-axis and y-axis
plt.xlabel("Median_Income")
plt.ylabel("Frequency")

# add a title to the plot
plt.title("Distribution of Median_Income")

#show the plot
plt.show()
```

```python
Observations:
    
>The histogram plot shows the frequency distribution of the median income values in the data set.
>The x-axis represents the median income values and the y-axis represents the frequency or count of those values.
>The plot shows that the majority of the median income values are between 0 and 4, with a peak around 2.5. 
>There are also some outliers with higher median income values above 8.
>The resulting histogram shows that the median income is not normally distributed, but rather it is skewed to the right. 
>The majority of the incomes are clustered around $20,000 to $70,000,but there are also some high-income outliers that 
make the distribution positively skewed.
```

```python
#9. Find the mean and median of the median income for the data set created in question 8.

# create a subset of the original dataframe where ocean_proximity is 'NEAR OCEAN'
near_ocean_df = df[df['ocean_proximity'] == 'NEAR OCEAN']

# calculate the mean of the median income for the subset
median_income_mean = near_ocean_df['median_income'].mean()

# calculate the median of the median income for the subset
median_income_median = near_ocean_df['median_income'].median()

# print the results
print("Mean of median income for Near Ocean dataset:", median_income_mean)
print("Median of median income for Near Ocean dataset:", median_income_median)

```

Observations:

indicating that the data may be relatively normally distributed.The mean of the median_income for the data set created in
question 8 is approximately 4, while the median of the median income is approximately 3.64. This suggests that the 
distribution of median income is slightly skewed to the right, as the mean is greater than the median. However, 
the difference between the mean and median is not very large, indicating that the skewness is relatively mild.

```python
#10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small.
#If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

# define a custom function to assign size categories based on total bedrooms
def assign_size(total_bedrooms):
    if total_bedrooms <= 200:
        return 'small'
    elif total_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

# create a new column  total_bedroom_size based on total bedrooms using the custom function
df['total_bedroom_size'] = df['total_bedrooms'].apply(assign_size)

# print the first 5 rows to verify the new column is created
print(df.head())
```

observations:
    The new column 'total_bedroom_size' has been added to the DataFrame with appropriate labels 'small', 'medium' and 'large'
based on the total number of bedrooms.
This will help in further analysis of the data based on the bedroom size categories.
